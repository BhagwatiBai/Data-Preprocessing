# Group by

```python
uber_df.groupby(['START*']).count()# how many rides were taken from each and every start place?
```

```python
uber_df.groupby('START*')['MILES*'].count()
```

```python
uber_df.groupby('START*')['MILES*'].agg(['mean']) # to find the average miles per start destination.
```

```python
uber_df.groupby('START*')['MILES*'].agg(['count','mean','sum','max','min'])
```

# Conditinally adding new column

```python
import numpy as np
```

```python
uber_df['MILES CATEGORY*'] = np.where(uber_df['MILES*']>10, "Long Trip", "Short Trip")

# the first value passed in np.where method is condition to be checked.
# the second  value passed in np.where method is the value to be populated when condition is satisfied
# the third  value passed in np.where method is the value to be populated when condition is not satisfied.
```

```python
uber_df
```

```python
uber_df['MILES CATEGORY*'].value_counts().plot.bar()
```

```python
uber_df.groupby(['MILES CATEGORY*']).agg(['count', 'max','min'])
```
